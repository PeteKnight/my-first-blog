```python
# import libraries and data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# Set the csv file path/name to be used
loadData = 'joinTable (3).csv' # file path/name ------------------------------------ POPULATE

batch_data = pd.read_csv(loadData)

batch_data.head()
```

```python
batch_data.shape
```

##Difference

```python
# Set the tile of the Plot/Graph
plot_title = 'Days' # Plot Title -------------------------------------- POPULATE

# Set the Data to be used in the x & y plots
x_data = 'difference' # x Data ---------------------------------------------- POPULATE
y_data = 'days' # y Data ---------------------------------------------- POPULATE

# Set the tiles to be used in the x & y plots
x_title = 'Difference' # x Label -------------------------------------------- POPULATE
y_title = 'Days' # y Label -------------------------------------------- POPULATE
```

```python
X = batch_data[[x_data]]
y = batch_data[y_data]


fig, ax = plt.subplots(figsize = (10, 4))

ax.scatter(X, y)

ax.set_title(plot_title)
ax.set_xlabel(x_title)
ax.set_ylabel(y_title)

plt.show()
```

```python
X = batch_data[[x_data]]
y = batch_data[y_data]


fig, ax = plt.subplots(figsize = (15, 4))

ax.scatter(X, y)

ax.set_title(plot_title)
ax.set_xlabel(x_title)
ax.set_ylabel(y_title)
plt.xlim(0, 5)
plt.show()
```
